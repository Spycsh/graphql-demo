```python
##########比对grpc和graphql##########
# encoding=utf-8
import matplotlib.pyplot as plt
from pylab import *                                 #支持中文
mpl.rcParams['font.sans-serif'] = ['SimHei']

names = ['100', '200', '300', '400', '500','600','700','800','900','1000']
x = range(len(names))
y=[244,553,762,975,1156,1327,1405,1639,1852,1962]
y1 = [52, 105, 141,175,205 ,237, 288, 317,375,432]
#plt.plot(x, y, 'ro-')
#plt.plot(x, y1, 'bo-')
#pl.xlim(-1, 11)  # 限定横轴的范围
#pl.ylim(-1, 110)  # 限定纵轴的范围
plt.plot(x, y, marker='o', mec='r', mfc='w',label=u'response time for grpc')
plt.plot(x, y1, marker='*', ms=10,label=u'response time for graphQL')
plt.legend()  # 让图例生效
plt.xticks(x, names, rotation=45)
plt.margins(0)
plt.subplots_adjust(bottom=0.15)
plt.xlabel(u"number of requests") #X轴标签
plt.ylabel("time(ms)") #Y轴标签
plt.title("Response time of grpc and graphQL") #标题

plt.show()
```

```python
##########比对graphql part attributes与含all attributes ##########
## 结果显示查询所有attributes整体比part attributes需要更多时间
# encoding=utf-8
import matplotlib.pyplot as plt
from pylab import *                                 #支持中文
mpl.rcParams['font.sans-serif'] = ['SimHei']

names = ['1', '2', '3', '4', '5','6','7','8','9','1']
x = range(len(names))
# 1000 次
y=[406,406,396,400,411,418,426,420,420,411]  # part data
y1 = [496,428,422,414,433,430,471,498,492,426]  # all data
#plt.plot(x, y, 'ro-')
#plt.plot(x, y1, 'bo-')
#pl.xlim(-1, 11)  # 限定横轴的范围
#pl.ylim(-1, 110)  # 限定纵轴的范围
plt.plot(x, y, marker='o', mec='r', mfc='w',label=u'response time for query of emloyees')
plt.plot(x, y1, marker='*', ms=10,label=u'response time for query of emplyees with address')
plt.legend()  # 让图例生效
plt.xticks(x, names, rotation=45)
plt.margins(0)
plt.subplots_adjust(bottom=0.15)
plt.xlabel(u"request index") #X轴标签
plt.ylabel("time(ms)") #Y轴标签
plt.title("Response time of query with or without all attributes") #标题

plt.show()
```

```python
##########比对graphql 条件查询单个employee与查询所有employees ##########
## 结果显示, 条件查询整体比查询所有employees要更耗时，且更加不稳定？可能原因是解析参数的过程
# encoding=utf-8
import matplotlib.pyplot as plt
from pylab import *                                 #支持中文
mpl.rcParams['font.sans-serif'] = ['SimHei']

names = ['1', '2', '3', '4', '5','6','7','8','9','1']
x = range(len(names))
# 1000 次
y=[427,461,415,445,401,455,456,501,476,446]  # employeeById
y1 = [416,469,525,415,568,459,559,525,428,452]  # allEmployees
#plt.plot(x, y, 'ro-')
#plt.plot(x, y1, 'bo-')
#pl.xlim(-1, 11)  # 限定横轴的范围
#pl.ylim(-1, 110)  # 限定纵轴的范围
plt.plot(x, y, marker='o', mec='r', mfc='w',label=u'response time for query of all emloyees')
plt.plot(x, y1, marker='*', ms=10,label=u'response time for query of one specific employes')
plt.legend()  # 让图例生效
plt.xticks(x, names, rotation=45)
plt.margins(0)
plt.subplots_adjust(bottom=0.15)
plt.xlabel(u"request index") #X轴标签
plt.ylabel("time(ms)") #Y轴标签
plt.title("Response time of query with conditional query or overall query") #标题

plt.show()
```
