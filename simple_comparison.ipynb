```python
##########比对grpc和graphql##########
# encoding=utf-8
import matplotlib.pyplot as plt
from pylab import *

mpl.rcParams['font.sans-serif'] = ['SimHei']


import numpy as np

def f_fit(x,y_fit):
    a,b=y_fit.tolist()
    print(y_fit.tolist())
    return a*x+b

def get_y_fitted(x, y):
    y_fit = np.polyfit(x,y,1)
    return f_fit(x, y_fit)

names = ['100', '200', '300', '400', '500','600','700','800','900','1000']
x = np.array(range(len(names)))
y=[54,58,84,120,143,168,202,221,263,307]
y1=[186,250,370,506,594,729,846,969,1091,1196]


y_fitted = get_y_fitted(x, y)
y1_fitted = get_y_fitted(x, y1)


plt.plot(x,y_fitted, mfc='w',label=u'response time for requesting all\n employee information in gRPC')
plt.plot(x,y1_fitted,color='orange', ms=10,label=u'response time for requesting all\n employee information in GraphQL')

plt.scatter(x,y,c='royalblue')#散点图

plt.scatter(x,y1,c='orange', marker='*')#散点图


# plt.plot(x, y, marker='o', mfc='w',label=u'response time for requesting all\n employee information in gRPC')
# plt.plot(x, y1, color='orange',marker='*', ms=10,label=u'response time for requesting all\n employee information in GraphQL')
plt.legend()  # 让图例生效
plt.xticks(x, names, rotation=45)
plt.margins(0)
plt.subplots_adjust(bottom=0.15)
plt.xlabel(u"number of requests") #X轴标签
plt.ylabel("time(ms)") #Y轴标签
plt.title("Response time for requesting all employee list using gRPC and GraphQL") #标题

plt.show()
```

```python
##########比对graphql part attributes与含all attributes ##########
## 结果显示查询所有attributes整体比part attributes需要更多时间
# encoding=utf-8
import matplotlib.pyplot as plt
from pylab import *                                 #支持中文
mpl.rcParams['font.sans-serif'] = ['SimHei']

names = ['1', '2', '3', '4', '5','6','7','8','9','1']
x = range(len(names))
# 1000 次
y=[406,406,396,400,411,418,426,420,420,411]  # part data
y1 = [496,428,422,414,433,430,471,498,492,426]  # all data
#plt.plot(x, y, 'ro-')
#plt.plot(x, y1, 'bo-')
#pl.xlim(-1, 11)  # 限定横轴的范围
#pl.ylim(-1, 110)  # 限定纵轴的范围
plt.plot(x, y, marker='o', mec='r', mfc='w',label=u'response time for query of emloyees')
plt.plot(x, y1, marker='*', ms=10,label=u'response time for query of emplyees with address')
plt.legend()  # 让图例生效
plt.xticks(x, names, rotation=45)
plt.margins(0)
plt.subplots_adjust(bottom=0.15)
plt.xlabel(u"request index") #X轴标签
plt.ylabel("time(ms)") #Y轴标签
plt.title("Response time of query with or without all attributes") #标题

plt.show()
```
